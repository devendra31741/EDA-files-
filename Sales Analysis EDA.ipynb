# Univariant Analysis

```python
def count_plot(column):
    plt.figure(figsize=(15,9))
    total=len(sales['Rating'])
    plot=sns.countplot(x=column,data=sales)
    plt.xlabel(f'{column}',fontsize=16)
    plt.ylabel("Number of Transaction")
    plt.title("Number of Transaction Classified by {}".format(column),fontsize=20)
    for p in plot.patches:
        plot.annotate('{:.1f}%'.format(100*p.get_height()/total),(p.get_x(),p.get_height()),fontsize=16)
    plot.yaxis.set_ticks(np.linspace(0,total,11))
    plot.set_yticklabels(map('{:.1f}%'.format,100*plot.yaxis.get_majorticklocs()/total))
```

```python
count_plot('Branch')
```

As we can see that there is not much difference in the overall transaction of the branch, however the branch A have done the maximum transaction 

```python
count_plot('City')
```

As we can see that there is not much difference in the overall transaction in the city, however the City A have done the maximum transaction

# Multi Variant Analysis

```python
#plot the mean and median when there is only one parameter 
#will show the mean and median of the graph
def dist_plot(column1):
    sns.displot(sales[column1],kde=True)
    plt.axvline(np.mean(sales[column1]),c='red',ls='--', label='mean')
    plt.axvline(x=np.percentile(sales[column1],25),c='green',ls='--',label='25-75 percentile')
    plt.axvline(x=np.percentile(sales[column1],75),c='green',ls='--')
    plt.legend(bbox_to_anchor=(1.05,1.0), loc="upper left")
    plt.xlabel(f'{column1}',fontsize=16)
    plt.title("Distribution of {}".format(column1),fontsize=20)
    plt.show()
```

```python
dist_plot('Rating')
```

Ratings were distributed evenly and there is not much difference. However the 6.5 rating is mostly given by the people, however the mean of all rating is 7

```python
sales['Time']=sales["Time"].astype('str')
sales['Hour']=sales['Time'].str[:2]
sales['Hour']=sales['Hour'].astype('int')
```

```python
sales.head()
```

```python
#defining a function to break a day into 6 sessions/periods using elif condition.
def f(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'                 #Creating sessions with conditions based on Hour Early morning, morning,noon,eve,night,late night.
    if (x > 8) and (x<=12):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'

sales['session'] = sales['Hour'].apply(f)               #Creating a session column based on Hour.
sales.head()
```

```python
sales_Session = sales.groupby(sales["session"])['Quantity'].sum().to_frame()
sales_Session=sales_Session.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
sns.barplot(x='session',y='Quantity',data=sales_Session, ax=ax)
```

Maximum sale during Noon and least on Morning
